```python
import pandas as pd
import itertools
```

```python
symbols= pd.read_csv('top500marketcap.csv')['Symbol']
```

```python


# Unique Combination Pairs for list of elements
def uniqueCombinations(list_elements):
    l = list(itertools.combinations(list_elements, 2))
    s = set(l)
    s= list(l)
    unique_pairs=[]
    for item in s:
        pair=item[0]+'_'+item[1]
        unique_pairs.append(pair)
    return unique_pairs
```

```python
s=uniqueCombinations(symbols)
```
