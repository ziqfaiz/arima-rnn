```python
import pandas as pd
import os
import re
```

```python
df = pd.read_csv('nasdaq_screener_1701174290288.csv')
df.head()
```

```python
df =df.nlargest(500,['Market Cap'])
```

```python
indexBrk = df[(df['Symbol'] == 'BRK/A') | (df['Symbol'] == 'BRK/B')].index
df.drop(indexBrk, inplace=True)
```

```python
df.to_csv('top500marketcap.csv',index=False)
```

```python
ten_year=[]
directory = "timeseries/"
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        name = re.search(r'/(.*?)_', f).group(1).split('/')[0]
        #print(name)
        temp_df = pd.read_csv(f)[["Date","Close"]]
        print(len(temp_df))
        if len(temp_df) == 2516:
            ten_year.append(name)
        del temp_df
```

```python
df=df[df.Symbol.isin(ten_year) == True]
```

```python
#pick 100 random stocks
random_list=df['Symbol'].sample(n=100, random_state=10)
```

```python
random_df=df[df.Symbol.isin(random_list) == True]
```

```python
random_df.to_csv('randomstocks.csv',index=False)
```
